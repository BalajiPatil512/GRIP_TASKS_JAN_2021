## Spliting the Dataset into Independent and Dependent Variables

```python
# Spliting the Datatset into Independent Variables

X = iris_data.iloc[:,1:5]
```

```python
#Checking the First, Five rows from Independent Variables

X.head()
```

```python
#Checking the shape of the Independent Variable

X.shape
```

```python
#Spliting the dataset into Dependent Variable

y = iris_data['Species']
```

```python
#Checking the first, Five rows from Dependent Variables

y.head()
```

```python
#Checking the shape of Dependent Variables

y.shape
```

# Spliting the Independent and Dependent Variables into Train and Test

```python
#Importing the Library and Spliting the data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
#Checking the Shape of X_train

X_train.shape
```

```python
#Checking the Shape of y_train

y_train.shape
```

```python
#Checking the Shape of X_test

X_test.shape
```

```python
#Checking the Shape of y_test

y_test.shape
```

# Applying the Decision Tree Algorithm

```python
#importing the library and fiting the X_train and y_train into Algorithm

from sklearn.tree import DecisionTreeClassifier
Dclassifier = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
#Predicting the test results

y_pred = Dclassifier.predict(X_test)
```

```python
#Test Results

y_pred
```

# Checking the Accuracy

```python
#Checking the Accuracy y_test vs y_pred

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn import tree
```

```python
i = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
d = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

# Visualising the Tree

```python
#Visualizing the Tree with the help of matplotlib

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5), dpi=350)
tree.plot_tree(Dclassifier, feature_names=i, class_names=d, filled=True);
```
